## Tests for `newrelic_sb_sdk.utils.validators`

## Imports

```python
from newrelic_sb_sdk.utils.validators import validate_user_key
```

## Tests

```python
class TestUtilsValidators:
    def test_validate_user_key_valid(self):
        assert validate_user_key("NRAK-123456789012345678901234567") is True

    def test_validate_user_key_invalid_prefix(self):
        assert validate_user_key("NRAL-123456789012345678901234567") is False

    def test_validate_user_key_invalid_length(self):
        assert validate_user_key("NRAK-12345678901234567") is False

    def test_validate_user_key_invalid_characters(self):
        assert validate_user_key("NRAK-12345678901234567890123456a") is False
```
