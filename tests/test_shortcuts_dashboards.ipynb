# Tests for `newrelic_sb_sdk.shortcuts.dashboards`

## Imports

```python
from unittest.mock import MagicMock, patch

import pytest
```

```python
from newrelic_sb_sdk.client import NewRelicClient
from newrelic_sb_sdk.graphql.objects import DashboardEntity, DashboardEntityOutline
from newrelic_sb_sdk.shortcuts.dashboards import get_all_dashboards, get_dashboard
```

## Tests

```python
class TestShortcutsDashboards:
    @patch("newrelic_sb_sdk.shortcuts.dashboards.NewRelicClient.execute")
    def test_get_all_dashboards(self, mock_execute):
        mock_response = MagicMock()
        mock_response.json.return_value = {
            "data": {
                "actor": {
                    "entitySearch": {
                        "count": 1,
                        "counts": [{"count": 1, "facet": "ACCOUNT_ID"}],
                        "results": {
                            "nextCursor": None,
                            "entities": [
                                {
                                    "__typename": "DashboardEntityOutline",
                                    "guid": "MjEyMzF8REFTSEJPQVJEfDAxMjM0",
                                    "name": "Test Dashboard",
                                    "account": {"id": 12345, "name": "Account"},
                                    "createdAt": "2024-01-01T00:00:00.000Z",
                                    "dashboardParentGuid": None,
                                    "entityType": "DASHBOARD_ENTITY",
                                    "firstIndexedAt": "2024-01-01T00:00:00.000Z",
                                    "owner": {"email": "test@example.com", "userId": 1},
                                    "permissions": "PRIVATE",
                                    "tags": [
                                        {"key": "isDashboardPage", "values": ["false"]}
                                    ],
                                    "type": "DASHBOARD",
                                    "updatedAt": "2024-01-01T00:00:00.000Z",
                                }
                            ],
                        },
                    }
                }
            }
        }
        mock_execute.return_value = mock_response

        client = NewRelicClient(new_relic_user_key="NRAK-123456789012345678901234567")
        data = get_all_dashboards(client=client)

        assert len(data) == 1
        assert isinstance(data[0], DashboardEntityOutline)
        assert data[0].guid == "MjEyMzF8REFTSEJPQVJEfDAxMjM0"

    @patch("newrelic_sb_sdk.shortcuts.dashboards.NewRelicClient.execute")
    def test_get_dashboard(self, mock_execute):
        mock_response = MagicMock()
        mock_response.json.return_value = {
            "data": {
                "actor": {
                    "entity": {
                        "__typename": "DashboardEntity",
                        "account": {"id": 12345, "name": "Account"},
                        "createdAt": "2024-01-01T00:00:00.000Z",
                        "dashboardParentGuid": None,
                        "description": "desc",
                        "domain": "APM",
                        "entityType": "DASHBOARD_ENTITY",
                        "firstIndexedAt": "2024-01-01T00:00:00.000Z",
                        "guid": "MjEyMzF8REFTSEJPQVJEfDAxMjM0",
                        "indexedAt": "2024-01-01T00:00:00.000Z",
                        "lastReportingChangeAt": "2024-01-01T00:00:00.000Z",
                        "name": "Test Dashboard",
                        "owner": {"email": "test@example.com", "userId": 1},
                        "permalink": "https://nr/",
                        "permissions": "PRIVATE",
                        "type": "DASHBOARD",
                        "updatedAt": "2024-01-01T00:00:00.000Z",
                        "pages": [
                            {
                                "__typename": "DashboardEntityPage",
                                "createdAt": "2024-01-01T00:00:00.000Z",
                                "description": "",
                                "guid": "p1",
                                "name": "Page 1",
                                "owner": {"email": "a", "userId": 1},
                                "updatedAt": "2024-01-01T00:00:00.000Z",
                                "widgets": [
                                    {
                                        "__typename": "DashboardEntityWidget",
                                        "id": "w1",
                                        "layout": {
                                            "column": 1,
                                            "height": 1,
                                            "row": 1,
                                            "width": 1,
                                        },
                                        "rawConfiguration": {},
                                        "title": "w1",
                                        "visualization": {"id": "v1"},
                                    }
                                ],
                            }
                        ],
                    }
                }
            }
        }
        mock_execute.return_value = mock_response

        client = NewRelicClient(new_relic_user_key="NRAK-123456789012345678901234567")
        data = get_dashboard(client=client, guid="MjEyMzF8REFTSEJPQVJEfDAxMjM0")

        assert isinstance(data, DashboardEntity)
        assert data.guid == "MjEyMzF8REFTSEJPQVJEfDAxMjM0"
        assert len(data.pages) == 1
```
